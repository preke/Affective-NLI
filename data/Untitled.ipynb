```python
import pandas as pd

df = pd.read_csv('CPED_A_VAD.tsv', sep='\t')
df.head()
```

```python
def get_uttr(row):
    ans = ''
    for i in range(len(row['new_dialog_state'])):
        if row['new_dialog_state'][i] == 0:
            pass
        else:
            ans += eval(row['sent'])[i] + ' '
    if ans == "":
        ans = '<pad>'
    return ans

def get_context(row):
    ans = ''
    for i in range(len(row['new_dialog_state'])):
        if row['new_dialog_state'][i] == 1:
            pass
        else:
            ans += eval(row['sent'])[i] + ' '
    if ans == "":
        ans = '<pad>'
    return ans


def get_sub_df(df, flow_len):
    sub_df = pd.DataFrame([])
    sub_df['labels'] = df['labels']
    sub_df['dialog_state'] = df['dialog_state'].apply(lambda x: eval(x)[:int(len(eval(x))*flow_len)])
    df['new_dialog_state'] = sub_df['dialog_state']
    sub_df['sent'] = df['sent'].apply(lambda x: eval(x)[:int(len(eval(x))*flow_len)])
    sub_df['utterance'] = df.apply(get_uttr, axis=1)
    sub_df['context'] = df.apply(get_context, axis=1)
    sub_df['Emoberta_softmax'] = df['Emoberta_softmax'].apply(lambda x: eval(x)[:int(len(eval(x))*flow_len)])
    sub_df['VAD_personality'] = df['VAD_personality']
    
    sub_df['sent'] = sub_df['sent'].apply(lambda x: ' '.join(x))
    
    return sub_df
for p in ['A', 'C', 'E', 'O', 'N']:
    df = pd.read_csv('CPED_'+p+'_VAD.tsv', sep='\t')
    for flow_len in [0.25, 0.75, 0.5]: 
        sub_df = get_sub_df(df, flow_len)
        print(sub_df.shape)
        print(sub_df[sub_df['sent'].isnull()])
        sub_df.to_csv('CPED_'+p+'_VAD_'+str(flow_len)+'.tsv', sep='\t', index=False)
```
